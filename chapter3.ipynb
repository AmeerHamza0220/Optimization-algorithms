```python
def bracket_minimum(f, x=0, s=1e-2, k=2):
    """  Find the minimum of a function f using the bracket method """
    a, ya = x, f(x)
    b, yb = a+s, f(a+s)
    if yb > ya:
        a, b = b, a
        ya, yb = yb, ya
        s = -s
    while True:
        c, yc = b+s, f(b+s)
        if yc > yb:
            return (a, c) if a < c else (c, a)
        a, ya, b, yb = b, yb, c, yc
        s *= k


def f(x): return x**2


minima = bracket_minimum(f)
print(minima)
```

```python
# plot f
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(-1, 1, 100)
y = f(x)
plt.plot(x, y)
# plot minima
plt.plot(minima[0], f(minima[0]), 'o')
```
